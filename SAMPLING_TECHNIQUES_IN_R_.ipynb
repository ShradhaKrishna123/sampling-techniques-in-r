```r
#Sampling techniques

#Creating a sample set(Step 1)

set.seed(100)
data<- data.frame(
    ID=1:100,
    Category=sample(c("A","B","C"),size=100,replace=TRUE),  #replace =TRUE , to repeat.
    Sales=round(runif(100,1000,20000),2), #2= 2 decimal points.
    Profit=round(runif(100,1000,50000),2),
    UnitsSold=sample(10:500,100,replace=TRUE),
    Score=round(runif(100,1,10),1)
)
head(data)
```

```r
# SIMPLE RANDOM SAMPLING

set.seed(123)
n<- 10
simple_random <-data[sample(nrow(data),n), ]
print('Simple Random Sampling:')
print(simple_random)
```

```r
#SYSTEMIC SAMPLING
set.seed(1)
N <- nrow(data)
n <- 10
k <- floor(N/n)
start <- sample(2:k , 1)
indices <- seq(start, by=k,length.out=n)
indices <- ((indices - 1)%% N) +1
systematic <- data[indices,]
print("SYSTEMATIC SAMPLING:")
print(systematic)
```

```r
#STRATIFIED DATA
set.seed(123)
prop <- 0.2
stratified <- data %>%
  group_by(Category)%>%
  slice_sample(prop=prop)%>%
  ungroup()
print(stratified)
```

```r
#CLUSTER SAMPLING
set.seed(1)
data$ClusterID <- sample(1:10, nrow(data), replace = TRUE)
chosen_clusters <- sample(unique(data$ClusterID), 3)
cluster_one_stage <- data %>% filter(ClusterID %in% chosen_clusters) # %>% = then do this %in% = matching operator.
print(cluster_one_stage)
```

```r
#NON PROBABILITY
#CONVENIENCE SAMPLING
convenience<-head(data,10)
print(convenience)
```

```r
#QUOTA SAMPLING
quotas <- data.frame(Category = c("A", "B", "C"), quota = c(5, 5, 5))

quota_sample <- do.call(rbind, lapply(seq_len(nrow(quotas)), function(i) {
  cat_i <- quotas$Category[i]
  q <- quotas$quota[i]
  data %>% filter(Category == cat_i) %>% head(q)
}))

print(quota_sample)
```

```r
# JUDGEMENTAL OR PURPOSIVE SAMPLING
purposive <- data %>% filter(Score > 8 & Sales > 15000)
print(purposive)
```

```r
# SNOWBALL SAMPLING
set.seed(123)
data$ID <- 1:nrow(data)
seeds <- sample(data$ID , 3)
snowball_ids <- seeds
max_waves <- 3
wave <- 1

while (wave <= max_waves){
  new_refs <- c()
  for (id in setdiff(snowball_ids,new_refs)){
    possible <- setdiff(data$ID , snowball_ids)
    k <- sample(0:2,1)
    if (length (possible)>0 && k>0){
      picks <- sample(possible,size=min(k,length(possible)))
      new_refs <- unique(c(new_refs,picks))
    }
  }
  if(length(new_refs)==0)break
    snowball_ids <- unique(c(snowball_ids,new_refs))
    wave <- wave+1
}
snowball<- data%>% filter (ID %in% snowball_ids)
print(snowball)
```
